```python
import pandas as pd
import geopandas as gp
from shapely import wkt

# Read the file as a CSV (which it currently is)
df = pd.read_csv(r"C:\Users\Anish\OneDrive\Desktop\minor\OUTPUT\Map_data.geojson")

# Convert the geometry text strings into actual shapes
df['geometry'] = df['geometry'].apply(wkt.loads)

# Convert to GeoDataFrame
gdf = gp.GeoDataFrame(df, geometry='geometry')

# Save as actual GeoJSON
gdf.to_file(r"C:\Users\Anish\OneDrive\Desktop\minor\OUTPUT\Map_data.geojson", driver='GeoJSON')
print("Successfully converted Map_data.geojson to valid GeoJSON format!")
```
